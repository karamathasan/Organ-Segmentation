# Model Install and data processessing

```python
model_name = "wholeBody_ct_segmentation"
download(name=model_name, bundle_dir=data_dir) # only download if necessary. No need to reinstall if you need perserve your config files
```

```python
model_path = os.path.join(data_dir, 'wholeBody_ct_segmentation', 'models', 'model_lowres.pt')
config_path = os.path.join(data_dir, 'wholeBody_ct_segmentation', 'configs', 'inference.json')
```

```python
config = ConfigParser()
config.read_config(config_path)
```

```python
preprocessing = config.get_parsed_content("preprocessing")
dataloader = config.get_parsed_content("dataloader") # data loader can be used if you want to preprocess in a more memory efficient manner
```

```python
model = config.get_parsed_content("network")
```

```python
model.load_state_dict(torch.load(model_path, map_location="cpu"))
model.eval();
```

```python
inferer = config.get_parsed_content("inferer")
```

```python
postprocessing = config.get_parsed_content("postprocessing")
```

```python
data = preprocessing({'image': CT_folder})

with torch.no_grad():
    data['pred'] = inferer(data['image'].unsqueeze(0), network=model)

data['pred'] = data['pred'][0]
data['image'] = data['image'][0]

data = postprocessing(data)
segmentation = torch.flip(data['pred'][0], dims=[2])
segmentation = segmentation.cpu().numpy()
```

# Segment

```python
image_loader = LoadImage(image_only=True)
CT = image_loader(CT_folder)

channel_transform = EnsureChannelFirst()
CT = channel_transform(CT)

orientation_transform = Orientation(axcodes=('LPS'))
CT = orientation_transform(CT)
```

```python
slice_idx = 256
CT_coronal_slice = CT[0, :, slice_idx].cpu().numpy()
segmentation_coronal_slice = segmentation[:,slice_idx]
```

```python
plt.subplots(1,2,figsize=(6,8))
plt.subplot(121)
plt.pcolormesh(CT_coronal_slice.T, cmap='Greys_r')
plt.axis('off')
plt.subplot(122)
plt.pcolormesh(segmentation_coronal_slice.T, cmap='nipy_spectral')
plt.axis('off')
plt.show()
```

```python
saggittal_slice_idx = 256
CT_saggittal_slice = CT[0, saggittal_slice_idx, :].cpu().numpy()
segmentation_saggittal_slice = segmentation[saggittal_slice_idx,:]
```

```python
plt.subplots(1,2,figsize=(6,8))
plt.subplot(121)
plt.pcolormesh(CT_saggittal_slice.T, cmap='Greys_r')
plt.axis('off')
plt.subplot(122)
plt.pcolormesh(segmentation_saggittal_slice.T, cmap='nipy_spectral')
plt.axis('off')
plt.show()
```

# Organ Volume Calculation

```python
def get_organ_volume(organ_channel):
  voxels = (segmentation == organ_channel).sum().item()
  vox_volume = np.prod(CT.meta['spacing']/10)
  organ_volumne = voxels * vox_volume
  return organ_volumne
```

```python
print("Bladder Volume: " + str(get_organ_volume(104)) + " cm^3")
print("Liver Volume: " + str(get_organ_volume(5)) + " cm^3")
print("Liver Volume: " + str(get_organ_volume(5)) + " cm^3")
```

```python
heart_myocardium_vol = get_organ_volume(44)
heart_atrium_left_vol = get_organ_volume(45)
heart_ventricle_left_vol = get_organ_volume(46)
heart_atrium_right_vol = get_organ_volume(47)
heart_ventricle_right_vol = get_organ_volume(48)

print("heart_myocardium Volume: " + str(heart_myocardium_vol) + " cm^3")
print("heart_atrium_left Volume: " + str(heart_atrium_left_vol) + " cm^3")
print("heart_ventricle_left Volume: " + str(heart_ventricle_left_vol) + " cm^3")
print("heart_atrium_right Volume: " + str(heart_atrium_right_vol) + " cm^3")
print("heart_ventricle_right Volume: " + str(heart_ventricle_right_vol) + " cm^3")

print("total heart volume: " + str(heart_myocardium_vol + heart_atrium_left_vol + heart_ventricle_left_vol + heart_atrium_right_vol + heart_ventricle_right_vol))
```
